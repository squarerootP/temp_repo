```python
from passlib.context import CryptContext

pwd_context = CryptContext(schemes=["argon2"], deprecated="auto")
def verify_password(plain_password, hashed_password):
    return pwd_context.verify(plain_password, hashed_password)

def get_password_hash(password):
    return pwd_context.hash(password)
```

```python
import os
from pathlib import Path

EXCLUDED_DIRS = {
    "__pycache__", ".git", ".idea", ".vscode",
    ".pytest_cache", ".mypy_cache", "dist", "build"
}

EXCLUDED_FILE_EXTS = {
    ".pyc", ".pyo", ".log", ".tmp", ".bak", ".swp"
}

def print_directory_tree(root_dir, indent="", exclude_hidden=True):
    """
    Print a directory tree structure starting from root_dir,
    excluding hidden and auto-generated files/folders.
    """
    root = Path(root_dir)

    # Get all items and sort (directories first)
    items = sorted(root.iterdir(), key=lambda x: (not x.is_dir(), x.name.lower()))

    for i, item in enumerate(items):
        # Skip hidden files/directories if exclude_hidden is True
        if exclude_hidden and item.name.startswith('.'):
            continue

        # Skip excluded directories
        if item.is_dir() and item.name in EXCLUDED_DIRS:
            continue

        # Skip excluded file extensions
        if item.is_file() and item.suffix in EXCLUDED_FILE_EXTS:
            continue

        # Choose connector symbol
        is_last = i == len(items) - 1
        connector = "└── " if is_last else "├── "
        print(f"{indent}{connector}{item.name}")

        # Recurse into subdirectories
        if item.is_dir():
            next_indent = indent + ("    " if is_last else "│   ")
            print_directory_tree(item, next_indent, exclude_hidden)

# Example usage
if __name__ == "__main__":
    directory = os.curdir
    print(f"Directory tree for: {os.path.abspath(directory)}\n")
    print_directory_tree(directory)
```
