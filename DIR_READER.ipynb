```python
from passlib.context import CryptContext

pwd_context = CryptContext(schemes=["argon2"], deprecated="auto")
def verify_password(plain_password, hashed_password):
    return pwd_context.verify(plain_password, hashed_password)

def get_password_hash(password):
    return pwd_context.hash(password)
```

```python
import os
from pathlib import Path

def print_directory_tree(root_dir, indent="", exclude_hidden=True):
    """
    Print a directory tree structure starting from root_dir.
    
    Args:
        root_dir (str): The root directory to start from
        indent (str): Current indentation (used in recursion)
        exclude_hidden (bool): Whether to exclude hidden files/directories
    """
    root = Path(root_dir)
    
    # Get all items in the directory
    items = list(root.iterdir())
    items.sort(key=lambda x: (not x.is_dir(), x.name.lower()))
    
    # Count of items to process
    count = len(items)
    
    for i, item in enumerate(items):
        # Skip hidden files/directories if exclude_hidden is True
        if exclude_hidden and item.name.startswith('.'):
            continue
        
        # Check if this is the last item at this level
        is_last = i == count - 1
        
        # Choose the appropriate connector symbols
        connector = "└── " if is_last else "├── "
        
        # Print the current item
        print(f"{indent}{connector}{item.name}")
        
        # If it's a directory, recursively print its contents
        if item.is_dir():
            # Adjust indentation for the next level
            next_indent = indent + ("    " if is_last else "│   ")
            print_directory_tree(item, next_indent, exclude_hidden)

# Example usage
if __name__ == "__main__":
    
    # Use command line argument if provided, otherwise use current directory
    directory = os.curdir
    
    print(f"Directory tree for: {os.path.abspath(directory)}")
    print_directory_tree(directory)
```
